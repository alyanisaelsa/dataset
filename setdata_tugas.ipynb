```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn import tree
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/alyanisaelsa/dataset/main/dataset.csv')
df
```

```python
columns = df.columns
df[columns] = df[columns].apply(lambda x: x.astype('category')) 
df[columns] = df[columns].apply(lambda x: x.cat.codes) 
df
```

```python
df_x = df.iloc[:, 0:11] #atribut
df_y = df.iloc[:,11] #kelas
df_y
```

```python
X = df_x.to_numpy()
y = df_y.to_numpy()
```

```python
gnb = GaussianNB()
model = gnb.fit(X, y) 
```

```python
# 19, private, 406078, Some-college, 10, Never-married, Adm-clerical, Own-child, white, Female, United-States
x_test = [[2,1,14,3,2,2,6,3,1,0,2]]

y_pred = model.predict(x_test)
y_pred
```
